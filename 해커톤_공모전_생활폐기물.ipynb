# 생활폐기물 처리 분석

```python
import pandas as pd
pd.read_csv('/content/폐기물_처리현황_생활폐기물_20240115225727.csv',encoding='cp949')
```

```python
data=pd.read_csv('/content/폐기물_처리현황_생활폐기물_20240115225727.csv'
,encoding='cp949',nrows=42,skiprows=[0,1])
data #전국 지역 단위만 추출
```

```python
data['폐기물 종류']=data['구분(3)']+" "+data['구분(4)']
data
```

```python
data=data.rename(columns={'소계':'발생량 소계'}) #열 이름 바꾸기
data
```

```python
data=data.drop(index=[0,1,2,10,19,41])
data #음식물류 폐기물 분리배출, 재활용 가능자원 분리배출, 종량제방식 등 혼합배출 방식의 폐기물 합계 데이터 삭제
```

```python
data['재활용률']=data['재활용']/data['발생량 소계']
data
```

```python
data_sorted=data.sort_values('재활용률',ascending=True) #오름차순, 뒤로 갈수록 높아짐
data_sorted
```

```python
data_last=data_sorted.head(10)
data_last #재활용률이 낮은 폐기물 상위 10개
```

```python
!pip install koreanize-matplotlib
import koreanize_matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(data_last['폐기물 종류'], data_last['재활용률'],
        color='skyblue',label='재활용률')
plt.title('소재별 재활용률')
plt.legend()
plt.xlabel('소재')
plt.ylabel('재활용률(재활용/발생량 소계)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
#재활용률이 가장 낮은 것은 불연성 폐유리류이다.
```

# 지역별 폐기물 처리현황 시각화

```python
import pandas as pd
pd.read_csv('/content/폐기물_처리현황_생활폐기물_20240115225727.csv',encoding='cp949')
```

```python
dataset=pd.read_csv('/content/폐기물_처리현황_생활폐기물_20240115225727.csv',encoding='cp949',skiprows=[0,1])
dataset
```

```python
dataset['폐기물 종류']=dataset['구분(3)']+" "+dataset['구분(4)']
dataset
```

```python
dataset=dataset.rename(columns={'소계':'발생량 소계'}) #열 이름 바꾸기
dataset
```

```python
dataset['재활용률']=dataset['재활용']/dataset['발생량 소계']
dataset
```

```python
dataset = dataset.drop(dataset[dataset['구분(3)'] == '소계'].index)
dataset
```

```python
dataset=dataset.drop(dataset[(dataset['구분(3)']=='가연성') & (dataset['구분(4)']=='소계')].index)
dataset
```

```python
dataset=dataset.drop(dataset[(dataset['구분(3)']=='불연성') & (dataset['구분(4)']=='소계')].index)
dataset
```

```python
!pip install koreanize-matplotlib
import koreanize_matplotlib
```

```python
import matplotlib.pyplot as plt


def make_graph():
  area = input("생활폐기물 재활용 현황을 알고 싶은 지역명을 입력해주세요: ")
  valid_areas = ['서울','부산','대구','인천','광주','대전','울산','세종','경기'
  ,'강원','충북','충남','전북','전남','경북','경남','제주']

  if area in valid_areas and len(area) == 2:
      print("올바른 지역명을 입력했습니다.")
      choice = dataset[dataset['구분(1)'] == area]

      plt.figure(figsize=(100, 50))

      plt.bar(choice['폐기물 종류'],choice['재활용'],label=area + " 재활용량")

      plt.title(area + " 재활용량",fontsize=100)
      plt.xlabel('소재',fontsize=80)
      plt.ylabel(area+"재활용량(ton)",fontsize=80)
      plt.xticks(rotation=45, ha='right',fontsize=50)
      plt.yticks(fontsize=50)

      plt.legend(fontsize=80)
      plt.show()

  else:
      print("지역명이 틀렸거나 두 글자의 시/도명이 아닙니다.")
      make_graph()


make_graph()
```

문제. 같이 첨부한 '2021년 서울시 자치구별 5대 범죄 발생 현황' 자료를 활용하여 아래 지시사항을 수행하시오. (3번째 행이 제목 행, 4번째 ‘소계’ 행은 제외)

1. 폭력 발생건수가 많은 상위 5개 자치구의 '폭력 발생' 건수를 65% 투명한 바차트로 나타내시오.
2. 위 1항목에 추가로 '폭력 검거' 건수를 역삼각형 마커를 사용한 빨간색 선차트로 나타내시오
3. 차트의 제목은 '본인이름_폭력 발생 및 검거 분석'으로 지정하시오. ex)홍길동_폭력 발생 및 검거 분석
4. 범례와 라벨을 표시하시오. (라벨 표시는 각각 '폭력발생건수' , '폭력검거건수' 로 지정)

```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('2021년 서울시 5대 범죄발생현황.csv', encoding='cp949', header=[3], skiprows=[4])

top_5_violence = a.sort_values('폭력_발생', ascending=False).head(5)

plt.bar(top_5_violence['자치구별'], top_5_violence['폭력_발생'], label='폭력_발생', alpha=0.65)

plt.show()
```

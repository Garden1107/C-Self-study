문제2. 국적별 관광 상용 자료 파일을 활용하여, 상용목적으로 가장 많이 방문한 국가순으로 정렬하는 전체코드를 작성하시오.
- 상용목적이 500건 이상인 데이터만 표시할 것
- 상용목적이 많은 순서대로 표시할 것 (내림차순으로 표시하는 코드 작성)
- 최종 결과가 들어갈 변수 이름은 ‘s2’로 지정

```python
import pandas as pd
a = pd.read_excel('국적별 관광 상용 자료.xlsx')
```

```python
a
```

```python
a.query('상용>=500')
b=a.query('상용>=500')
```

```python
b.sort_values('상용', ascending=False)
```

```python
s2=b.sort_values('상용', ascending=False)
```

```python
import pandas as pd
import folium
df = pd.read_excel('/content/서울시 공영 주차장.xlsx')
map = folium.Map(location=[37.4673761,126.8925673], zoom_start=13)
for n,i in df.iterrows():
		mk = folium.Marker([ i['Latitude'] , i['Longitude'] ] , tooltip= i['주차장명'] )
		mk.add_to(map)
map
```
